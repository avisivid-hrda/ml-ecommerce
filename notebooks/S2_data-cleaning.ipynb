| Original Column Name    | Standardized Column Name       |
|-------------------------|-------------------------------|
| *old_col_name*        | *`new_col_name`*        |

```python
import pandas as pd

# Load main dataset
df_main = pd.read_csv("../data/raw/online_shoppers_intention.csv")

# Define the Renaming Dictionary
rename_columns = {
    "old_col_name": "new_col_name",
    "old_col_name": "new_col_name",
    "old_col_name": "new_col_name",
    "old_col_name": "new_col_name"
}

# Rename columns using the rename Dictionary
df_main.rename(columns=rename_columns, inplace=True)

# Save updated dataset into a new CSV file
df_main.to_csv("../data/processed/online_shoppers_intention_standardized.csv", index=False)

# Display Standardized DataFrame
print("Standardized DataFrame:")
display(df_main.head())

# Display the list of the standardidized column names in df_main
print("Standardized Column Names in df_main:")
display(df_main.columns.tolist())
display(df_main.info())
```
